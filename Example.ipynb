```python
som=df[df['country']=='Somalia']
som
```

```python
k=df[df['country']=='Kenya']
k
```

```python
sns.set_style('whitegrid')
plt.plot(som.year, som.population, 'y', marker='*', lw=1)
plt.title('popilation of somalia since 1954')
plt.ylabel('pouplation')
plt.xlabel('year')
plt.show()
```

```python
sns.set_style('whitegrid')
plt.plot(som.year, som.population, 'y', marker='*', lw=1, label='somalia')
plt.plot(k.year, k.population, 'g', marker='*', lw=1, label='Kenya')
plt.title('popilations Som Vs K since 1954')
plt.ylabel('pouplation')
plt.xlabel('year')
plt.legend()
plt.show()
```
